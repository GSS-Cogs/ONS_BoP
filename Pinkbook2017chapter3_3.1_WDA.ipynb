```python
tab = loadxlstabs('PB2017_3.1.xls')
```

```python
tab = tab[0]
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D8').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
Code = tab.excel_ref('C8').expand(DOWN).is_not_blank()
```

```python
Year = tab.excel_ref('D5').expand(RIGHT).is_not_blank()
```

```python
Transactions = tab.excel_ref('B').expand(DOWN).by_index([7,24,41]) - tab.excel_ref('B57').expand(DOWN)
```

```python
Services = tab.excel_ref('B').expand(DOWN).is_not_blank().is_not_whitespace() - Transactions - tab.excel_ref('B2').expand(UP)
```

```python
Currency = tab.excel_ref('W3')
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'Year', DIRECTLY,ABOVE),
            HDim(Code,'Cdid',DIRECTLY,LEFT),
            HDimConst('Currency', 'Pounds(millions)'),            
            HDim(Services, 'Type of Services', DIRECTLY, LEFT),
            HDim(Transactions,'Transactions',CLOSEST, ABOVE)           
    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
```

```python
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table.drop('DATAMARKER', axis=1)
```

```python
new_table['TIME'] = new_table['Year'].map(lambda cell:cell.replace('.0', ''))
new_table
```

```python
new_table = new_table.drop('Year', axis=1)
```

```python
print(len(new_table))
new_table
```

```python
new_table['Type of Services'] = new_table['Type of Services'].str.strip()
```

```python
new_table['TIME'] = new_table['TIME'].str.strip()
```

```python
new_table['GEOG'] = new_table['GEOG'].str.strip()
```

```python
new_table['Currency'] = new_table['Currency'].str.strip()
```

```python
new_table['Cdid'] = new_table['Cdid'].str.strip()
```

```python
new_table['Transactions'] = new_table['Transactions'].str.strip()
```
