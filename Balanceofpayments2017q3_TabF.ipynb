```python
tab = loadxlstabs('BoP_TabF.xls')
```

```python
tab = tab[0]
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D9').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
savepreviewhtml(observations)
```

```python
Code = tab.excel_ref('C9').expand(DOWN).is_not_blank()
```

```python
savepreviewhtml(Code)
```

```python
Year = tab.excel_ref('D5').expand(RIGHT).is_not_blank()
Quarter = tab.excel_ref('D6').expand(RIGHT)
```

```python
savepreviewhtml(Year)
```

```python
savepreviewhtml(Quarter)
```

```python
Trade = tab.excel_ref('B').expand(DOWN).by_index([8,23,38])
```

```python
savepreviewhtml(Trade)
```

```python
Services = tab.excel_ref('B').is_not_blank().is_not_whitespace() - Trade
```

```python
Services = Services - tab.excel_ref('B2').expand(UP)
```

```python
savepreviewhtml(Services)
```

```python
Currency = tab.excel_ref('N4')
```

```python
savepreviewhtml(Currency)
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'TIME',DIRECTLY,ABOVE),
            HDim(Quarter,'Quarter',DIRECTLY,ABOVE),
            HDim(Code,'Cdid',DIRECTLY,LEFT),
            HDimConst('Currency', 'Pounds(millions)'),            
            HDim(Services,'Services',CLOSEST, ABOVE),
            HDim(Trade,'Trade',CLOSEST,ABOVE)
           
    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table.drop('TIMEUNIT', axis=1)
```
