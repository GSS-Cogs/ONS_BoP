```python
tab = loadxlstabs('PB2017_3.9.xls')
```

```python
tab = tab[0]
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D9').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
Code = tab.excel_ref('C9').expand(DOWN).is_not_blank()
```

```python
Year = tab.excel_ref('D4').expand(RIGHT).is_not_blank()
```

```python
Commerce = tab.excel_ref('B').expand(DOWN).by_index([6,18,31]) - tab.excel_ref('B32').expand(DOWN)
```

```python
Revenue = tab.excel_ref('B').expand(DOWN).is_not_blank().is_not_whitespace() - Commerce 
```

```python
Currency = tab.excel_ref('W3')
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'TIME', DIRECTLY,ABOVE),
            HDim(Code,'Cdid',DIRECTLY,LEFT),
            HDimConst('Currency', 'Pounds(millions)'),            
            HDim(Revenue, 'Revenue', DIRECTLY, LEFT),
            HDim(Commerce,'Commerce', CLOSEST, ABOVE)    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
```

```python
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table.drop('TIMEUNIT', axis=1)
new_table = new_table.drop('DATAMARKER', axis=1)
```

```python
new_table['Revenue'] = new_table['Revenue'].map(lambda cell:cell.replace('of which:', ''))
```

```python
new_table['Revenue'] = new_table['Revenue'].str.rstrip('23')
```

```python
print(len(new_table))
new_table
```

```python
new_table['TIME'] = new_table['TIME'].str.strip()
new_table['GEOG'] = new_table['GEOG'].str.strip()
new_table['Currency'] = new_table['Currency'].str.strip()
new_table['Cdid'] = new_table['Cdid'].str.strip()
```

```python
new_table['Commerce'] = new_table['Commerce'].str.strip()
```

```python
new_table['Revenue'] = new_table['Revenue'].str.strip()
```
