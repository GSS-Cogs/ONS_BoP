```python
tab = loadxlstabs('PB2017_3.2.xls')
```

```python
tab = tab[0]
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D10').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
Code = tab.excel_ref('C10').expand(DOWN).is_not_blank()
```

```python
Year = tab.excel_ref('D4').expand(RIGHT).is_not_blank()
```

```python
Commerce = tab.excel_ref('B').expand(DOWN).by_index([6,67,129]) - tab.excel_ref('B184').expand(DOWN)
```

```python
Modeoftransport = tab.excel_ref('B').expand(DOWN).by_index([8,32,46,52,58,60,62,64,69,98,104,108,112,114,118,
                                                            120,122,124,126,131,135,142,149,152,153,155,165,171,177,179,181,183])
                                                                
```

```python
Modeoftransport = Modeoftransport - tab.excel_ref('B184').expand(DOWN)
```

```python
Services = tab.excel_ref('B').expand(DOWN).by_index([10,17,21,26,28,30,34,35,37,38,40,42,44,64,71,78,82,86,88,90,94,96,98,106,
                                                     126,131,155,163,183]) - tab.excel_ref('B184').expand(DOWN)
```

```python
Typeoftransport = tab.excel_ref('B').expand(DOWN).is_not_blank().is_not_whitespace() -Services - Modeoftransport - tab.excel_ref('B2').expand(UP)
```

```python
Currency = tab.excel_ref('W2')
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'Year', DIRECTLY,ABOVE),
            HDim(Code,'Code Identifier',DIRECTLY,LEFT),
            HDimConst('Currency', 'Pounds(millions)'),            
            HDim(Typeoftransport, 'Revenue', DIRECTLY, LEFT),
            HDim(Services, 'Services', CLOSEST, ABOVE),
            HDim(Modeoftransport, 'Mode of Transport', CLOSEST, ABOVE),
            HDim(Commerce, 'Commerce', CLOSEST, ABOVE)
    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
```

```python
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table.drop('DATAMARKER', axis=1)
```

```python
new_table['TIME'] = new_table['Year'].map(lambda cell:cell.replace('.0', ''))
new_table
```

```python
new_table = new_table.drop('Year', axis=1)
```

```python
print(len(new_table))
new_table
```

```python
new_table['Mode of Transport'] = new_table['Mode of Transport'].map(lambda x: x.rstrip('12'))
```
